```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("FedEx dataset.csv")

print("âœ… Dataset Loaded Successfully!\n")
print("Shape of the data:", df.shape)
print("Columns:\n", df.columns.tolist())

df.head()
```

```python
# Display the last few rows of the dataset
df.tail()
```

```python
# Show concise summary of the DataFrame
df.info()
```

```python
df.describe(include='all')
```

```python
# Count missing values in each column
data_dict = pd.DataFrame({
    'Column Name': df.columns,
    'Data Type': df.dtypes.astype(str),
    'Non-Null Count': df.notnull().sum().values,
    'Null Count': df.isnull().sum().values,
    'Null Percentage (%)': (df.isnull().mean() * 100).round(2),
    'Description': [''] * len(df.columns)
})

data_dict
```

```python
desc = {
    'ID': 'Unique identifier for each record',
    'Project Code': 'Internal code representing the project',
    'PQ #': 'Pre-qualification request number',
    'PO / SO #': 'Purchase Order or Sales Order number',
    'ASN/DN #': 'Advance Shipment Notice or Delivery Note number',
    'Country': 'Destination country for the shipment',
    'Managed By': 'Team managing the shipment (e.g., PMO)',
    'Fulfill Via': 'Method used to fulfill the order (e.g., Direct Drop or From RDC)',
    'Vendor INCO Term': 'International commercial terms between buyer and vendor',
    'Shipment Mode': 'Transportation mode (e.g., Air, Sea)',
    'PQ First Sent to Client Date': 'Date when PQ was first sent to the client',
    'PO Sent to Vendor Date': 'Date when Purchase Order was sent to vendor',
    'Scheduled Delivery Date': 'Scheduled delivery date for the order',
    'Delivered to Client Date': 'Actual delivery date to the client',
    'Delivery Recorded Date': 'Date the delivery was recorded in system',
    'Product Group': 'General group or category of the product',
    'Sub Classification': 'More specific product classification',
    'Vendor': 'Name of the vendor supplying the product',
    'Item Description': 'Description of the item or product',
    'Molecule/Test Type': 'Molecule name or test associated with the product',
    'Brand': 'Product brand',
    'Dosage': 'Dosage information for pharmaceuticals',
    'Dosage Form': 'Form of medication (e.g., tablet, injection)',
    'Unit of Measure (Per Pack)': 'Units per pack (e.g., 10 tablets)',
    'Line Item Quantity': 'Quantity ordered per line item',
    'Line Item Value': 'Total cost or value of the line item in USD',
    'Pack Price': 'Price per pack in USD',
    'Unit Price': 'Price per individual unit in USD',
    'Manufacturing Site': 'Location of the manufacturing facility',
    'First Line Designation': 'Indicates if this is a priority product',
    'Weight (Kilograms)': 'Weight of the shipment in kilograms',
    'Freight Cost (USD)': 'Cost of shipping in USD',
    'Line Item Insurance (USD)': 'Insurance cost for the line item in USD'
}

data_dict['Description'] = data_dict['Column Name'].map(desc)
data_dict
```

```python
# Count missing values in each column
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'].replace(r'[^\d.]','',regex=True), errors='coerce')
df['Freight Cost (USD)']    = pd.to_numeric(df['Freight Cost (USD)'].replace(r'[^\d.]','',regex=True), errors='coerce')

date_cols = [
    'PQ First Sent to Client Date',
    'PO Sent to Vendor Date',
    'Scheduled Delivery Date',
    'Delivered to Client Date',
    'Delivery Recorded Date'
]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='%d-%b-%Y', errors='coerce')

df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for correlation
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Correlation matrix
corr_matrix = numeric_cols.corr()

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

The heatmap above shows correlation between numeric columns.
We can observe if there are any strong relationships (positive/negative), such as between weight and freight cost, or PO sent and delivery time.


```python
# Convert string dates to datetime format
# Convert date columns (if not already)
df["PO Sent to Vendor Date"] = pd.to_datetime(df["PO Sent to Vendor Date"])
df["Scheduled Delivery Date"] = pd.to_datetime(df["Scheduled Delivery Date"])

# Create Lead Time column
df["Lead Time (Days)"] = (df["Scheduled Delivery Date"] - df["PO Sent to Vendor Date"]).dt.days

# Trend plot
plt.figure(figsize=(10, 5))
sns.histplot(df["Lead Time (Days)"], bins=20, kde=True)
plt.title("Distribution of Lead Time")
plt.xlabel("Lead Time (Days)")
plt.ylabel("Number of Shipments")
plt.show()
```

This chart shows the distribution of lead time between PO sent and scheduled delivery.
We can identify how long shipments usually take and detect any abnormal delays.


```python
# Visualize data using plots
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Weight (Kilograms)", y="Freight Cost (USD)", hue="Shipment Mode")
plt.title("Freight Cost vs Weight by Shipment Mode")
plt.xlabel("Weight (Kilograms)")
plt.ylabel("Freight Cost (USD)")
plt.show()
```

This scatter plot shows the relation between shipment weight and freight cost.
As expected, heavier shipments usually cost more, and cost also varies with shipment mode.


```python
# Convert string dates to datetime format
# Convert delivery date column
df["Delivered to Client Date"] = pd.to_datetime(df["Delivered to Client Date"])

# Flag for on-time
df["On Time"] = df["Delivered to Client Date"] <= df["Scheduled Delivery Date"]

# Countplot
plt.figure(figsize=(7, 5))
sns.countplot(x="On Time", data=df)
plt.title("On-Time vs Delayed Deliveries")
plt.xlabel("Delivered On Time")
plt.ylabel("Count")
plt.xticks([0, 1], ["Delayed", "On Time"])
plt.show()
```

We created an 'On Time' flag and visualized how many shipments were on time versus delayed.
This gives insights into logistics efficiency.


```python
# Visualize data using plots
plt.figure(figsize=(7, 5))
sns.countplot(x="Shipment Mode", data=df)
plt.title("Shipment Mode Distribution")
plt.xlabel("Shipment Mode")
plt.ylabel("Number of Shipments")
plt.show()
```

This bar chart shows the most frequently used shipment modes in the dataset.


```python
# Convert string dates to datetime format
# Convert relevant date columns to datetime
df['PO Sent to Vendor Date'] = pd.to_datetime(df['PO Sent to Vendor Date'], errors='coerce')
df['Scheduled Delivery Date'] = pd.to_datetime(df['Scheduled Delivery Date'], errors='coerce')

# Calculate lead time in days
df['Lead Time (Days)'] = (df['Scheduled Delivery Date'] - df['PO Sent to Vendor Date']).dt.days

# Display result
df['Lead Time (Days)'].describe()
```

```python
# Calculate freight cost per kilogram
# Ensure data is numeric
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')

# Calculate Freight per KG
df['Freight per KG'] = df['Freight Cost (USD)'] / df['Weight (Kilograms)']

# Display result
df['Freight per KG'].describe()
```

```python
# Convert string dates to datetime format
# Convert delivery dates
df['Delivered to Client Date'] = pd.to_datetime(df['Delivered to Client Date'], errors='coerce')

# Create on-time flag
df['On Time'] = df['Delivered to Client Date'] <= df['Scheduled Delivery Date']

# Calculate percentage
on_time_percentage = df['On Time'].mean() * 100
print(f"On-Time Delivery Percentage: {on_time_percentage:.2f}%")
```

```python
# Get value counts of a categorical column
# Shipment volume by mode
shipment_volume = df['Shipment Mode'].value_counts()

# Average cost per shipment mode
avg_cost_by_mode = df.groupby('Shipment Mode')['Freight Cost (USD)'].mean()

# Average lead time by shipment mode
avg_lead_time = df.groupby('Shipment Mode')['Lead Time (Days)'].mean()

# Display results
print("Shipment Volume:\n", shipment_volume)
print("\nAverage Cost by Mode:\n", avg_cost_by_mode)
print("\nAverage Lead Time by Mode:\n", avg_lead_time)
```

```python
# Display the first few rows of the dataset
# Top 5 vendors by frequency
top_vendors = df['Vendor'].value_counts().head()

# Top 5 destinations (countries) by shipment volume
top_destinations = df['Country'].value_counts().head()

# Top 5 vendors by total freight cost
top_costly_vendors = df.groupby('Vendor')['Freight Cost (USD)'].sum().sort_values(ascending=False).head()

# Display the results
print("Top 5 Vendors (by frequency):\n", top_vendors)
print("\nTop 5 Destinations (by volume):\n", top_destinations)
print("\nTop 5 Vendors by Total Freight Cost:\n", top_costly_vendors)
```

```python
# Visualize data using plots
plt.figure(figsize=(8,5))
sns.barplot(x=top_vendors.values, y=top_vendors.index, color='skyblue')  # removed palette, added color
plt.title("Top 5 Vendors by Frequency")
plt.xlabel("Number of Shipments")
plt.ylabel("Vendor")
plt.tight_layout()
plt.show()
```

```python
# Get value counts of a categorical column
shipment_mode_counts = df['Shipment Mode'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(shipment_mode_counts, labels=shipment_mode_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Shipment Mode Distribution")
plt.axis('equal')
plt.show()
```

```python
# Convert string dates to datetime format
# Convert PO Date if not already datetime
df['PO Sent to Vendor Date'] = pd.to_datetime(df['PO Sent to Vendor Date'])

# Group by month
monthly_avg_freight = df.groupby(df['PO Sent to Vendor Date'].dt.to_period('M'))['Freight per KG'].mean()

# Line plot
monthly_avg_freight.plot(kind='line', marker='o', figsize=(10,5), title='Avg Freight per KG Over Time')
plt.ylabel('Freight per KG')
plt.xlabel('Month')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Calculate lead time in days
plt.figure(figsize=(8,5))
sns.histplot(df['Lead Time (Days)'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Lead Time (Days)")
plt.xlabel("Lead Time (Days)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Calculate lead time in days
plt.figure(figsize=(10,6))
corr = df[['Line Item Value', 'Freight Cost (USD)', 'Weight (Kilograms)', 'Lead Time (Days)', 'Freight per KG']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
